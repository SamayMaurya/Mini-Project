```python
import pandas as pd
import numpy as np
```

```python
train_df=pd.read_csv('/kaggle/input/icr-identify-age-related-conditions/train.csv')
train_df1=train_df
geeks_df=pd.read_csv('/kaggle/input/icr-identify-age-related-conditions/greeks.csv')
```

```python
train_df1=pd.get_dummies(train_df1,columns=['EJ'],drop_first=True)
train_df1
```

```python
train_df1['BQ'].fillna(train_df1['BQ'].mean(),inplace=True) 
train_df1['DU'].fillna(train_df1['DU'].mean(),inplace=True)
train_df1['FL'].fillna(train_df1['FL'].mean(),inplace=True)
train_df1['CB'].fillna(train_df1['CB'].mean(),inplace=True)
train_df1['CC'].fillna(train_df1['CC'].mean(),inplace=True)
train_df1['FC'].fillna(train_df1['FC'].mean(),inplace=True)
train_df1['FS'].fillna(train_df1['FS'].mean(),inplace=True)
train_df1['GL'].fillna(train_df1['GL'].mean(),inplace=True)
train_df1['EL'].fillna(train_df1['EL'].mean(),inplace=True)
```

```python
train_df1.isnull().sum()
```

```python
train_df1.keys()
```

```python
train_df1=train_df1.drop(['CB', 'CC', 'CR', 'CS', 'CU',
       'CW ', 'DA', 'DE', 'DH',  'DL', 'DN',
       'EE', 'EG','EP', 'EU', 'FI', 'GE', 'GF','GL','Id','AH','AY','BR','CF','BD '],axis=1)
```

```python
y=train_df1['Class']
```

```python
x=train_df1.drop(['Class'],axis=1)
x
```

```python
x.corr(method='kendall')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=12)
x_train_res, y_train_res = sm.fit_resample(x_train,y_train)
y_train_res
```

```python
x_train_res
```

```python
X_combined = np.vstack((x_train, x_train_res))
y_combined = np.hstack((y_train, y_train_res))
X_combined
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming you have your features (X) and labels (y)
# Replace X and y with your actual data



# Create a Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
clf.fit(X_combined, y_combined)

# Make predictions on the testing data
y_pred = clf.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Define the number of folds
k_folds = 10  # You can adjust the number of folds based on your requirements

# Create a KFold object
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Perform K-fold cross-validation
cross_val_results = cross_val_score(clf, x_test, y_test, cv=kf, scoring='accuracy')

# Print the accuracy for each fold
print("Cross-validation results (Accuracy):", cross_val_results)

# Print the mean accuracy across all folds
print("Mean Accuracy:", cross_val_results.mean())
```
